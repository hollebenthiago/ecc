```python
import numpy as np
from IPython.display import Audio
```

```python
# Generate a sound
framerate = 44100
t = np.linspace(0,5,framerate*5)
data = np.sin(2*np.pi*220*t) + np.sin(2*np.pi*224*t)
Audio(data,rate=framerate)
```

```python
Audio(np.random.permutation(data), rate=framerate)
# Audio(data, rate=framerate)
```
